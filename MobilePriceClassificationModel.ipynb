# Spliting data set as training and testing

```python
import sklearn.model_selection as sk
from sklearn import preprocessing
```

```python
y = data_file.price_range
x = data_file.drop('price_range', axis=1)
x.head(5)
```

```python
x_train, x_test, y_train, y_test = sk.train_test_split(x, y, test_size=0.25)
x_train = preprocessing.normalize(x_train)
x_train.shape
```

# Creating the ANN

```python
print(" +----------------------------+----------------------------+")
print(" |      Hyper Parameter       |   Binary Classification    |")
print(" +----------------------------+----------------------------+")
print(" | # input neurons            | One per input feature      |")
print(" | # hidden layers            | Typically 1 to 5           |")
print(" | # neurons per hidden layer | Typically 10 to 100        |")
print(" | # output neurons           | 1 per prediction dimension |")
print(" | Hidden activation          | ReLU, Tanh, sigmoid        |")
print(" | Output layer activation    | Sigmoid                    |")
print(" | Loss function              | Binary Cross Entropy       |")
print(" +----------------------------+----------------------------+")
print()
print(" +-----------------------------------+----------------------------+")
print(" |          Hyper Parameter          | Multiclass Classification  |")
print(" +-----------------------------------+----------------------------+")
print(" | # input neurons                   | One per input feature      |")
print(" | # hidden layers                   | Typically 1 to 5           |")
print(" | # neurons per hidden layer        | Typically 10 to 100        |")
print(" | # output neurons                  | 1 per prediction dimension |")
print(" | Hidden activation                 | ReLU, Tanh, sigmoid        |")
print(" | Output layer activation           | Softmax                    |")
print(" | Loss function                     | 'Categorical Cross Entropy |")
print(" | Sparse Categorical Cross Entropy' |                            |")
print(" +-----------------------------------+----------------------------+")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# model = tf.keras.models.Sequential()
# model.add(Dense(20, activation='tanh', input_shape=(20,)))
# model.add(Dense(8, activation='tanh'))
# model.add(Dense(4, activation='tanh'))

model = Sequential() 
model.add(Dense(units=8, activation='relu', input_dim=20)) 
model.add(Dense(units=4, activation='relu')) 
model.add(Dense(units=1, activation='sigmoid'))
```

# Compiling ANN

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='sgd',
    metrics=['accuracy']
)
model.summary()
```

# Fitting and training the model

```python
model.fit(x_train, y_train.values,validation_data=(x_test, y_test), batch_size=32, epochs=100)
```

# Saving weights

```python
model.save_weights('model_weights.h5')
# model.save('model_weights_01.h5')
```
